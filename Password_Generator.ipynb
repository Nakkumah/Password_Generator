# Password Generator
This notebook contains a simple Python script to generate a random password using words, numbers, and symbols.


```python
import random
import string

# List of words to use in the password
word_list = ['apple', 'banana', 'cherry', 'date', 'elderberry', 'fig', 'grape', 'honeydew']

# Function to generate a random password
def generate_password(length=12):
    # Randomly select words
    words = random.choices(word_list, k=2)
    
    # Random numbers and symbols
    numbers = ''.join(random.choices(string.digits, k=2))
    symbols = ''.join(random.choices(string.punctuation, k=2))
    
    # Combine words, numbers, and symbols
    password_elements = words + [numbers, symbols]
    random.shuffle(password_elements)
    
    # Combine all elements into a single string
    password = ''.join(password_elements)
    
    # Trim to the desired length if necessary
    if len(password) > length:
        password = password[:length]
    
    return password

# Generate a password and print it
password = generate_password(12)
print(f"Generated password: {password}")
```

## Explanation of the Code
- Import the `random` and `string` modules for generating random elements.
- Define a function `generate_password` that creates a password using words, numbers, and symbols.
- The password is generated by selecting random words, digits, and punctuation symbols.
- Finally, call the function and print the generated password.
